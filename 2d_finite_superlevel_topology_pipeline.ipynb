This notebook is a reproducible entry point for `finite_superlevel_topology_pipeline.py`. It calls the same script used for command-line replication so that notebook and script outputs remain mutually consistent.

The default notebook run executes the synthetic structural benchmark, the Wisconsin breast-cancer semi-synthetic benchmark, and the randomized JOBS II application. The optional eICU analysis is skipped unless a local eICU directory is supplied.

The implementation estimates smoothed two-dimensional outcome densities, evaluates finite density-superlevel filtrations, records Betti-0, Betti-1, Euler, and sliced Euler signatures, and reports covariate balance, overlap, permutation, trimming, bootstrap, and sensitivity diagnostics. These diagnostics document empirical stability; they do not prove conditional exchangeability or any other identifying assumption.

## Scientific Scope

The synthetic and breast-cancer semi-synthetic analyses are causal benchmarks because the potential outcomes are generated and therefore known. These benchmarks allow direct comparison between observed or weighted factual estimators and known-potential-outcome finite topological signatures.

The JOBS II analysis is a randomized empirical application. It supports randomized treatment-group comparisons, but it is not a structural hidden-confounding benchmark because unit-level potential outcomes are not observed.

The optional eICU analysis is observational. It is useful as a large-scale descriptive stress test of the finite-superlevel estimator, but it should not be interpreted as evidence that causal identification assumptions hold. Balance, overlap, permutation, trimming, and sensitivity diagnostics are reported as design diagnostics only.
